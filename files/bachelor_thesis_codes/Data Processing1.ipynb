```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

```python
eno = pd.read_csv('ENO_Date_CSV.csv')
eno
```

```python
yu = pd.read_csv('yuyichen.csv')
yu
```

```python
gu = pd.read_csv('gu.csv')
gu
```

```python
eno2 = eno[['Unnamed: 0','month']]
pd1 = pd.merge(eno2,gu,on='Unnamed: 0')
pd1
```

```python
yu2 = yu[['Unnamed: 0','block']]
yu2
```

```python
pd2 = pd.merge(yu2,pd1,on='Unnamed: 0')
pd2
```

```python
tiao = pd.read_csv('Features-sparse matrix.csv')
tiao
```

```python
pd3 = pd.merge(pd2,tiao,on='Unnamed: 0')
pd3
```

```python
pd3
```

```python
pd3['label'] = 0
```

```python
for i in range(0,len(pd3)):
    if pd3['interest_level'][i] == 'low':
        pd3['label'][i]=0
    elif pd3['interest_level'][i] == 'medium':
        pd3['label'][i]=1
    elif pd3['interest_level'][i] == 'high':
        pd3['label'][i]=2
```

```python
pd3
```

```python
del pd3['Unnamed: 0']
```

```python
pd3['Brooklyn']=0
pd3['Manhattan']=0
pd3['Queens']=0
pd3['The Bronx']=0
pd3['Staten Island']=0

for i in range(0,len(pd3)):
    if pd3['block'][i] == 'Brooklyn':
        pd3['Brooklyn'][i]=1
    elif pd3['block'][i] == 'Manhattan':
        pd3['Manhattan'][i]=1
    elif pd3['block'][i] == 'Queens':
        pd3['Queens'][i]=1
    elif pd3['block'][i] == 'The Bronx':
        pd3['The Bronx'][i]=1
    elif pd3['block'][i] == 'Staten Island':
        pd3['Staten Island'][i]=1
```

```python
pd3
```

```python
pd3.to_csv('Integrated Version.csv')
```

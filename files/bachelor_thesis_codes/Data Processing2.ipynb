```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

test=pd.read_json("test.json")
train=pd.read_json("train.json")
```

```python
train1 = train.reset_index()
train1
```

```python
train1.describe()
```

## Street Code

```python
v = train1['street_address'].value_counts()
pdn = pd.DataFrame(v)
pdn.index.name = 'street_address'
pdn.rename(columns={'street_address':'street_code'},inplace = True)
pdn
```

```python
pd1 = pd.merge(train1,pdn,on='street_address')
pd1
```

```python
i = pd1['street_code'].value_counts()
j = pd.DataFrame(i)
j.index.name = 'x'
h = j.reset_index()
h.loc[h['x'] == 174]#检验
```

## Building Code

```python
v1 = pd1['building_id'].value_counts()
pdn1 = pd.DataFrame(v1)
pdn1.index.name = 'building_id'
pdn1.rename(columns={'building_id':'building_code'},inplace = True)
pdn1
```

```python
pd2 = pd.merge(pd1,pdn1,on='building_id')
pd2
```

## Manager Code

```python
v2 = pd2['manager_id'].value_counts()
pdn2 = pd.DataFrame(v2)
pdn2.index.name = 'manager_id'
pdn2.rename(columns={'manager_id':'manager_code'},inplace = True)
pdn2
```

```python
pd3 = pd.merge(pd2,pdn2,on='manager_id')
pd3.sort_values(by='index',ascending=True, inplace=True)
pd3 = pd3.reset_index(drop=True)
pd3
```

```python
pd3.to_csv('gu.csv')
```
